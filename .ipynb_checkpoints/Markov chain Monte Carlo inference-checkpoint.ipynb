```python
import numpy as np
import math
from scipy.stats import multivariate_normal
from scipy.stats import norm
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr 
plt.style.use('ggplot')
```

```python
def proposal_q(x_l_1,stepsize):
    """
    Gaussian proposal for Random Walk Metropolis-Hastings algorithm
    
    :numpy x_l_1: previous step
    :param stepzise: stepsize of the Gaussian (variance)
    
    return
    :numpy x_cand: candidate
    """
    return np.random.normal(x_l_1,np.sqrt(stepsize))
```

```python
def ex_1(x):
    return np.prod(np.array(norm.pdf(x))) 
```

```python
def mh(p_star, param_init, num_samples=5000, stepsize=1.0):
    """
    Function to compute samples using the Random Walk Metropolis-Hastings 
    algorithm as e.g. given in Algorithm 27.3 of Barber (2012) 
    
    inputs
    :function p_star: a function on θ that is proportional to the density of interest p∗(θ)
    :list param_init: the initial sample — a value for θ from where the Markov chain starts
    :param num_samples: the number of samples to generate
    :param stepsize: a hyperparameter specifying the variance 
                     of the Gaussian proposal distribution q

    
    return
    :numpy samples:array of the samples
    """
    samples = np.zeros((num_samples, len(param_init)))
    x = param_init   
    

    for i in range(num_samples):
        
        x_cand = [proposal_q(x_i,stepsize) 
                             for x_i in x]
        
        a = p_star(x_cand) / p_star(x)
        if a >= 1:
            x = x_cand
        else:
            if np.random.rand() < a:
                x = x_cand
        samples[i] = x      
    
    return samples
```

```python
def mh_warm(p_star, param_init, num_samples=5000, stepsize=1.0,warm_up=0):
    """
    Function to compute samples using the Random Walk Metropolis-Hastings 
    algorithm as e.g. given in Algorithm 27.3 of Barber (2012) 
    
    inputs
    :function p_star: a function on θ that is proportional to the density of interest p∗(θ)
    :list param_init: the initial sample — a value for θ from where the Markov chain starts
    :param num_samples: the number of samples to generate
    :param stepsize: a hyperparameter specifying the variance 
                     of the Gaussian proposal distribution q
    :param warm_up: number of warm-up steps. If warm_up > 0 the algorithms will run
                    for num_samples + warm_up times and will return num_samples samples
    
    return
    :numpy samples:array of the samples
    """
    num_samples = num_samples + warm_up
    samples = np.zeros((num_samples, len(param_init)))
    x = param_init   
    

    for i in range(num_samples):
        
        x_cand = [proposal_q(x_i,stepsize) 
                             for x_i in x]
        
        a = p_star(x_cand) / p_star(x)
        if a >= 1:
            x = x_cand
        else:
            if np.random.rand() < a:
                x = x_cand
        samples[i] = x      
    
    samples = samples[warm_up:]
    return samples
```

```python
param_init = [0,0]
p_star = ex_1
samples = mh(p_star,param_init)
```

## Samples from p(x, y) = N (x; 0, 1)N (y; 0, 1)

## Bayesian Poisson regression model

```python
data = np.loadtxt("q3_poisson.txt")
```

```python
data_x = data[0]
data_y = data[1]

sum_ = data_x*data_y
sum_ = sum_.sum()
y_sum = data_y.sum()
```

```python
def ex_3(x):
    """
    x[0] alpha
    x[1] beta
    """
    global data_x
    global data_y
    global sum_
    global y_sum
    
    x_ = data_x*x[0] + x[1]
    x_exp = np.exp(x_)
    inside_exp = (-x[0]**2 - x[1]**2)/200 + x[0]*sum_ + x[1]*y_sum - x_exp.sum()
    
    return np.exp(inside_exp)
```

## Bayesian Poisson regression model sampling

```python
param_init = [0,0]
p_star = ex_3
samples = mh_warm(p_star,param_init,num_samples=5000, stepsize=1.0,warm_up=1000)
```
