# Rejection sampling

```python
import numpy as np
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
import math
```

```python
def rejection_sampling(n,p_):
    """
    Implementation of rejection sampling
    
    :param n: number of samples to draw
    :function p_: function that x_i ~ p_(x)
    """
    samples_accepted = []
    samples_rejected = []
    while len(samples_accepted) < n:
        y_i = scipy.stats.uniform.rvs(0,1,size=1)
        x_i = scipy.stats.uniform.rvs(0,3,size=1)
        if y_i < p_(x_i):
            samples_accepted.append(tuple((x_i,y_i)))
        else:
            samples_rejected.append(tuple((x_i,y_i)))
            
    return np.asarray(samples_accepted),np.asarray(samples_rejected)  
```

```python
def exponential(x_i):
    return math.exp(-x_i)
```

```python
accepted,rejected = rejection_sampling(1000,exponential)
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.scatter(x=accepted[:,0],y=accepted[:,1],label='Accepted')
ax.scatter(x=rejected[:,0],y=rejected[:,1],label='Rejected')
ax.legend()
plt.xlabel('x')
plt.ylabel('p(x)')
```
